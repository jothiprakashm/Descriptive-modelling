```python
import pandas as pd
import numpy as np


def one():    
    xls_file = pd.ExcelFile('Energy Indicators.xls')
    energy = xls_file.parse(skiprows=17,skip_footer=(38))
    energy = energy[['Unnamed: 1','Petajoules','Gigajoules','%']]
    
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
  
    energy[['Energy Supply', 'Energy Supply per Capita', '% Renewable']] =  energy[['Energy Supply', 'Energy Supply per Capita', '% Renewable']].replace('...',np.NaN).apply(pd.to_numeric)
    
    energy['Energy Supply'] = energy['Energy Supply']*1000000
   
    energy['Country'] = energy['Country'].replace({"Republic of Korea": "South Korea", "United States of America": "United States",
                                                   "United Kingdom of Great Britain and Northern Ireland": "United Kingdom", 
                                                   "China, Hong Kong Special Administrative Region": "Hong Kong"})

    energy['Country'] = energy['Country'].str.replace(r" \(.*\)","")
   
    GDP = pd.read_csv('world_bank.csv', skiprows=4)
    
    GDP = GDP[['Country Name','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']]
    GDP['Country Name'] = GDP['Country Name'].replace({"Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran", 
                                             "Hong Kong SAR, China": "Hong Kong"})
 
    GDP.columns = ['Country','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
    
    #GDP.head()
    ScimEn= pd.read_excel('scimagojr-3.xlsx')
    ScimEn_top = ScimEn[:15]
    
    ScimEn_Energy = pd.merge(ScimEn_top, energy, how='inner', left_on='Country', right_on='Country')
    df_joint = pd.merge(ScimEn_Energy, GDP, how='inner', left_on='Country', right_on='Country')
    df_joint = df_joint.set_index('Country')
    return df_joint

one()
```

```python
def three():
    Top15 = one()
    years=['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
    avgGDP= (Top15[years].mean(axis=1).sort_values(ascending=False))

    return avgGDP

three()
```

```python
def four():
    Top15 = one()
    years=['2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
    Top15['avgGDP']=(Top15[years].mean(axis=1).sort_values(ascending=False))
    Top15.sort_values(ascending=False, by= 'avgGDP',inplace= True)
    
    return abs(Top15.iloc[5]['2019']-Top15.iloc[5]['2009'])

four()
```

```python
def six():
    Top15 = one()
    a= Top15.sort_values(ascending=False, by= '% Renewable').iloc[0]
    return (a.name,a['% Renewable'])

six()
```

```python
def seven():
    Top15 = one()
    a=Top15['Self-citations']/Top15['Citations']
    Top15['citations_ratio']=a
    Ratio_rank=Top15.sort_values(ascending=False,by= 'citations_ratio').iloc[0]
    return (Ratio_rank.name,Ratio_rank['citations_ratio'])
seven()
```

```python
def eight():
    Top15 = one()
    peo_app= Top15['Energy Supply']/Top15['Energy Supply per Capita']
    Top15['energy_ratio']=peo_app
    App = Top15.sort_values(ascending=False, by='energy_ratio').iloc[2]
    
    return App.name

eight()
```

```python
def nine():
    Top15 = one()
    Top15['Estimate Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    Top15['Citable Documents per Person'] = Top15['Citable documents'] / Top15['Estimate Population']
    Top15['Citable Documents per Person']=np.float64(Top15['Citable Documents per Person'])
    data = Top15[['Citable Documents per Person','Energy Supply per Capita']]
    return data['Citable Documents per Person'].corr(data['Energy Supply per Capita'])
nine()
```

```python
def ten():
    Top15 = one()
    mid= Top15['% Renewable'].median()
    Top15['HighRenew'] = Top15['% Renewable']>= mid
    Top15['HighRenew'] = Top15['HighRenew'].apply(lambda x: 1 if x else 0) 
    Top15.sort_values(ascending=True,by= 'Rank')
    return Top15['HighRenew']

ten()
```

```python
def eleven():
    Top15 = one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    groups = pd.DataFrame(columns = ['size', 'sum', 'mean', 'std'])
    Top15['Estimate Population'] = Top15['Energy Supply'] / Top15['Energy Supply per Capita']
    for group, frame in Top15.groupby(ContinentDict):
        groups.loc[group] = [len(frame), frame['Estimate Population'].sum(),frame['Estimate Population'].mean(),frame['Estimate Population'].std()]
    return groups

eleven()
```

```python
def twelve():
    Top15 = one()
    Top15 = one()
    ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}
    Top15= Top15.reset_index()
    Top15['Continent'] = [ContinentDict[country] for country in Top15['Country']]
    Top15['bins']= pd.cut(Top15['% Renewable'],5)
    return Top15.groupby(['Continent','bins']).size()

twelve()
```

```python
def thirteen():
    Top15 = one()
    Top15['PopEst'] = (Top15['Energy Supply'] / Top15['Energy Supply per Capita']).astype(float)
    return  Top15['PopEst'].apply(lambda x: '{0:,}'.format(x))

thirteen()
```

```python
def plot():
    import matplotlib as plt
    %matplotlib inline
    Top15 = one()
    ax = Top15.plot(x='Rank', y='% Renewable', kind='scatter', 
                    c=['#e41a1c','#377eb8','#e41a1c','#4daf4a','#4daf4a','#377eb8','#4daf4a','#e41a1c',
                       '#4daf4a','#e41a1c','#4daf4a','#4daf4a','#e41a1c','#dede00','#ff7f00'], 
                    xticks=range(1,16), s=6*Top15['2019']/10**10, alpha=.75, figsize=[16,6]);

    for i, txt in enumerate(Top15.index):
        ax.annotate(txt, [Top15['Rank'][i], Top15['% Renewable'][i]], ha='center')

    print("This is a bubble chart showing % Renewable vs. Rank. The size of the bubble corresponds to the countries' \
2019 GDP, and the color corresponds to the continent.")
plot()    
```
